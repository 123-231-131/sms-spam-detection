## Data Preprocessing
## --> Lower Case
## --> Tokenization
## --> Removing Special characters
## --> Removing stop words and punctutaion
## --> Stemming

```python
from nltk.corpus import stopwords
nltk.download('stopwords')
stopwords.words('english')
```

```python
import string
string.punctuation
```

```python
def transform_text(text):
    text = text.lower()
    text = nltk.word_tokenize(text)
    y = []
    for i in text:
        if i.isalnum():
            y.append(i)

    text = y[:]
    y.clear()

    for i in text:
        if i not in stopwords.words('english') and i not in string.punctuation:
            y.append(i)

    text = y[:]
    y.clear()

    for i in text:
        y.append(ps.stem(i))
        
            
    return " ".join(y)
```

```python
transform_text("Dancing Machine How??")
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
ps.stem('dancing')
```

## Top 30 words in ham and spam

```python
spam_corpus = []
for msg in df[df['target'] == 1]['transformed_text'].tolist():
    for word in msg.split():
        spam_corpus.append(word)
```

```python
len(spam_corpus)
```

```python
from collections import Counter
# Create a DataFrame from Counter's most_common(30) result
df_most_common = pd.DataFrame(Counter(spam_corpus).most_common(30), columns=['word', 'count'])

# Use sns.barplot with x and y arguments
sns.barplot(x='word', y='count', data=df_most_common)
plt.xticks(rotation = 'vertical')
plt.show()
```

```python
ham_corpus = []
for msg in df[df['target'] == 0]['transformed_text'].tolist():
    for word in msg.split():
        ham_corpus.append(word)
```

```python
len(ham_corpus)
```

```python
df_most_common1 = pd.DataFrame(Counter(ham_corpus).most_common(30), columns=['word', 'count'])

# Use sns.barplot with x and y arguments
sns.barplot(x='word', y='count', data=df_most_common1)
plt.xticks(rotation = 'vertical')
plt.show()
```
